```python
import pandas as pd
import seaborn as sns

sns.set_theme()
df = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/penguins.csv')
print(df[df['species'] == 'Gentoo'])
df['species'].unique()
print("Average Weight Gentoo: " ,df[df['species'] == 'Gentoo']['body_mass_g'].mean())
print("Average Weight Chinstrap: " ,df[df['species'] == 'Chinstrap']['body_mass_g'].mean())
print("Average Weigth Gentoo Males: " ,df[(df['species'] == 'Gentoo') & (df['sex'] == "MALE")]['body_mass_g'].mean())
print("Average Weigth Chinstrap Males: " ,df[(df['species'] == 'Chinstrap') & (df['sex'] == "MALE")]['body_mass_g'].mean())

df.groupby('sex', as_index=False)['body_mass_g'].mean()
df.groupby(['sex','species'])['body_mass_g'].mean()

#sns.boxplot(data=df, x='body_mass_g', y='sex')
ax = sns.scatterplot(data=df, x="body_mass_g", y='flipper_length_mm', hue="species")
#ax.set(xlim=(0,7000))
```
